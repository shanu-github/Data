```python
from project_common_utils.automl_module.autosklearn_module import autosklearn_regression
from project_common_utils.automl_module.autosklearn_module import autosklearn_classification
import pandas as pd
import autosklearn
```

# Regression

```python
rdata = pd.read_csv('train_icr.csv')
rdata.head()
```

This Method runs the autosklearn algorithm and returns the results

**:param num_col**: (int) number of numerical columns\
**:param internal_eval_metric**: (autosklearn object) internal evaluation metric (r2, mse, etc.)\
**:param total_time**: (int) total desired time for running the autosklearn algorithm\
**:param per_model_time**: (int) time allocated to each model in autosklearn internally\
**:param cpu_cores**: (int) number of cpu cores to be used\
**:param resample**: (string) resampling strategy to be used (non-cv, cv)\
**:param data_name**: (string) assign a name to the data used\
**:param filename**: (string) data file name\
**:param split_strategy**: (string) random or stratified sampling criteria\
**:param testsize**: (float) test size to used\
**:param seed**: (int) seed to be used for reproducibility\
**:param target_col**: (string) target column to be mentioned\
**:param cat_col_names**: (string) categorical column names (optional, only if categorical columns are present)\
**:param stratify_col**: (string array) column to be used for stratified sampling (optional)\
**:param cv_fold**: (int) if cv is used, specify the cv folds (optional)

**:return**: returns model results and a excel file having predictions on test data

```python
results = autosklearn_regression(num_col = 21,
                                 internal_eval_metric = autosklearn.metrics.r2,
                                 total_time = 900,
                                 per_model_time = None,
                                 cpu_cores = None,
                                 resample = 'non_cv',
                                 data_name = 'ICR',
                                 filename = 'train_icr.csv',
                                 split_strategy = 'random',
                                 testsize = 0.3,
                                 seed = 42,
                                 target_col = 'Q_INCA')
```

```python
cdata = pd.read_csv('pcb_data.csv')
cdata.head()
```

This Method runs the autosklearn algorithm and returns the results

**:param num_col**: (int) number of numerical columns\
**:param internal_eval_metric**: (autosklearn object) internal evaluation metric (r2, mse, etc.)\
**:param total_time**: (int) total desired time for running the autosklearn algorithm\
**:param per_model_time**: (int) time allocated to each model in autosklearn internally\
**:param cpu_cores**: (int) number of cpu cores to be used\
**:param resample**: (string) resampling strategy to be used (non-cv, cv)\
**:param data_name**: (string) assign a name to the data used\
**:param filename**: (string) data file name\
**:param split_strategy**: (string) random or stratified sampling criteria\
**:param testsize**: (float) test size to used\
**:param seed**: (int) seed to be used for reproducibility\
**:param target_col**: (string) target column to be mentioned\
**:param num_class**: (string) binary or multi-class classification to be specified\
**:param sample_strategy**: (string) upsample/downsample/nosample strategy fro binary classification\
**:param cat_col_names**: (string) categorical column names (optional, only if categorical columns are present)\
**:param stratify_col**: (string array) column to be used for stratified sampling (optional)\
**:param cv_fold**: (int) if cv is used, specify the cv folds (optional)

**:return**: returns model results and a xlsx file having predictions on test data

```python
results = autosklearn_classification(num_col = 44,
                                     internal_eval_metric = autosklearn.metrics.f1_weighted,
                                     total_time = 300,
                                     per_model_time = None,
                                     cpu_cores = None,
                                     resample = 'non-cv',
                                     data_name = 'PCB',
                                     filename = 'pcb_data.csv',
                                     split_strategy = 'stratified',
                                     testsize = 0.3,
                                     seed = 42,
                                     target_col = 'Product class',
                                     num_class = 'multi',
                                     sample_strategy = 'nosample',
                                     cat_col_names = ['has_cooling_layer_a','has_cooling_layer_b','is_via','is_viap'],
                                     stratify_col = 'Product class')
```
